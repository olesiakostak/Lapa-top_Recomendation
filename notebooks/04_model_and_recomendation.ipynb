```python
similarity_matrix = cosine_similarity(scaled_df)
similarity_df = pd.DataFrame(similarity_matrix, columns=scaled_df.index, index=scaled_df.index)
similarity_df.head(2)
```

## Function that displays top N laptops similar to given one 

```python
def recomedate_similar_laptops(laptop_name, similarity_df, top_n=5): 
    if laptop_name not in similarity_df.index:
        return "Laptop name is not in given data base"
    
    most_similar = similarity_df[laptop_name].sort_values(ascending=False)[1:top_n+1]
    return pd.DataFrame(most_similar).transpose()
```
