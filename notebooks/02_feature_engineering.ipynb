```python
import pandas as pd
import re
```

```python
df = pd.read_csv('../data/laptop_cleaned.csv')
```

```python
df.head()
```

## One-Hot Encoding for 'OS' column

```python
df['OS'].value_counts()
```

```python
df = pd.get_dummies(df, columns=['OS'], dtype=int)
```

## Handling 'Display' column

```python
df['Display'].unique()
```

```python
df['Display'] = df['Display'].str.replace('\u2009', ' ')
```

```python
df['Screen Size'] = df['Display'].str.extract('(^\d+\.?\d+)')
```

```python
df['Touch Screen'] = pd.Series(1 if 'Touch Screen' in x else 0 for x in df['Display'])
```

```python
resolution = df['Display'].str.extract('(\d+)\s*[Xx]\s*(\d+)')
df['Total_Pixels'] = resolution[0].astype(float) * resolution[1].astype(float)
```

```python
df = df.drop(['Display'], axis=1)
```

```python
df.info()
```

```python
df['Screen Size'] = df['Screen Size'].astype(float)
```

```python
df.head()
```

```python
df = df.dropna()
```

## Handling 'Graphics' column

```python
df['Graphics'].unique()
```

```python
df['Graphics'] = df['Graphics'].str.replace('\u2009', ' ')
df['Graphics'] = df['Graphics'].str.replace('\u200e', '')
```

```python
df['VRAM'] = df['Graphics'].str.extract('(^\d+)')
df['VRAM'] = df['VRAM'].fillna(0)
df['VRAM'] = df['VRAM'].astype(int)
```

```python
def get_gpu_brand(text):
    text = text.lower()
    if 'amd' in text: return 'AMD'
    if 'intel' in text: return 'Intel'
    if 'nvidia' in text: return 'NVIDIA'
    if 'apple' in text or 'mac' in text or 'core gpu' in text: return 'Apple'
    return 'Other'

df['GPU brand'] = df['Graphics'].apply(get_gpu_brand)
```

```python
df = df.drop('Graphics', axis=1)
```

```python
df = pd.get_dummies(data=df, columns=['GPU brand'], dtype=int)
df.info()
```

## Handling 'Core' column

```python
df['Core'].unique()
```

```python
df['Thread'] = df['Core'].str.split(',', expand=True)[1]
df['Core'] = df['Core'].str.split(',', expand=True)[0]
```

```python
df['Thread'] = df['Thread'].str.extract('(\d+)')
df['Thread'] = df['Thread'].astype(float)
```

```python
df['Core'].value_counts()
```

```python
def get_cpu_brand(text):
    if 'intel' in text or 'celeron' in text or 'pentium' in text: return 'Intel'
    if 'amd' in text or 'ryzen' in text or 'athlon' in text: return 'AMD'
    if 'apple' in text or 'm1' in text or 'm2' in text or 'm3' in text: return 'Apple'
    return 'Other'

df['CPU_Brand'] = df['Generation'].apply(get_cpu_brand)
```

```python
def transform_core_data(text):
    val = re.search('(\d+)', str(text))
    if not val:
        text = text.lower()
        if 'dual' in text: return 2
        if 'quad' in text: return 4
        if 'hexa' in text: return 6
        if 'octa' in text: return 8
    return val.group(0)

df['Core'] = df['Core'].apply(transform_core_data)
```

```python
df['Core'] = df['Core'].astype(int)
```

```python
df['Thread'] = df['Thread'].fillna(df.groupby('Core')['Thread'].transform('median'))
df.info()
```

```python
df[df['Thread'].isnull()]
```

```python
df.groupby('Core')['Thread'].median()
```

```python
df['Thread'] = df['Thread'].fillna(14.0)
```

## Handling 'Generation' column

```python
df['Generation'].unique()
```

```python
df['Generation'] = df['Generation'].str.replace('\u200e', '')
df['Generation'] = df['Generation'].str.replace('\u200b', '')
df['Generation'] = df['Generation'].str.lower()
```

```python
def get_cpu_brand(text):
    if 'intel' in text or 'celeron' in text or 'pentium' in text: return 'Intel'
    if 'amd' in text or 'ryzen' in text or 'athlon' in text: return 'AMD'
    if 'apple' in text or 'm1' in text or 'm2' in text or 'm3' in text: return 'Apple'
    return 'Other'

df['CPU_Brand'] = df['Generation'].apply(get_cpu_brand)
```

```python
def get_cpu_tier(text):
    if 'i9' in text or 'ryzen 9' in text or 'ultra 9' in text: return 'Tier 9'
    if 'i7' in text or 'ryzen 7' in text or 'core 7' in text or 'ultra 7' in text: return 'Tier 7'
    if 'i5' in text or 'ryzen 5' in text or 'core 5' in text or 'ultra 5' in text: return 'Tier 5'
    if 'i3' in text or 'ryzen 3' in text or 'core 3' in text or 'ultra 3' in text: return 'Tier 3'
    if 'm1' in text or 'm2' in text or 'm3' in text: return 'Apple Silicon'
    return 'Basic/Other' 

df['Generation'] = df['Generation'].apply(get_cpu_tier)
```

```python
df = pd.get_dummies(df, columns=['Generation'], dtype=int)
df = pd.get_dummies(df, columns=['CPU_Brand'], dtype=int)
```

```python
df.info()
```

```python
df.set_index('Model', inplace=True)
```

```python
df.to_csv('..\data\laptop_featured.csv', index=True)
```
